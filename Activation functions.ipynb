## Sigmoid : output will be between 0 to 1

```python
import math
def sigmoid_fun(a):
    return 1 / (1 + math.exp(-a))
```

```python
print(sigmoid_fun(100))
print(sigmoid_fun(1))
print(sigmoid_fun(0))
print(sigmoid_fun(-100))
```

## tanh - output between +1 and -1

```python
def tanh_fun(a):
  return (math.exp(a) - math.exp(-a)) / (math.exp(a) + math.exp(-a))
```

```python
print(tanh_fun(70))
print(tanh_fun(-87))
```

## reLU - for positive values output will be same value and for negative value output will be 0.



```python
def relu(a):
    return max(0,a)
```

```python
print(relu(67))
print(relu(-67))
```

## Leaky ReLU - Postive input output will be same value  . For negative input , output will be  0.1 *a


```python
def leaky_relu_fun(a):
    return max(0.1*a,a)
```

```python
print(leaky_relu_fun(98))
print(leaky_relu_fun(-98))
```
